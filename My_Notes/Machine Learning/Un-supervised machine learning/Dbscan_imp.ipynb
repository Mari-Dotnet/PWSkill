### Dbscan clustring

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=make_moons(n_samples=250,noise=0.30)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
dbscan=DBSCAN(eps=0.3)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
plt.scatter(X_scaled[:,0],X_scaled[:,1],c=dbscan.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silscore=[]
for k in range(5,15):
    dbscan=DBSCAN(eps=0.3,min_samples=k)
    dbscan.fit(X_scaled)
    score=silhouette_score(X_scaled,dbscan.labels_)
    silscore.append(score)
```

```python
plt.plot(range(5,15),silscore,marker='o')
plt.xticks(range(2,11))
plt.show()
```

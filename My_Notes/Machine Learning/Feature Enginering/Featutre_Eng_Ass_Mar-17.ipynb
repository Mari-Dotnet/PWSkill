##### Q2: List down techniques used to handle missing data. Give an example of each with python code.
- mean value imputation
- median value imputation
- mode value imputation

```python
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.isnull().sum()
```

```python
## age is missing values we can update the missing valuses in mean values
df['age_mean']=df['age'].fillna(df.age.mean())
```

```python
df.isnull().sum()
```

```python
## update the median value
df['age_meadian']=df['age'].fillna(df.age.median())
```

```python
##update the mode values
df['embark_town_mode']=df['embark_town'].fillna(df['embark_town'].mode())
```

#### Q5: What is data Augmentation? Explain SMOTE.
- Data augmentation is a technique used to artificially increase the size of a dataset by generating new examples from existing data. 
- SMOTE is specifically designed for imbalanced datasets and works by generating synthetic examples of the minority class, which are then added to the dataset. The algorithm works by identifying a minority sample and selecting one or more of its k nearest neighbors. A new sample is then generated by interpolating between the minority sample and one of its neighbors.

#### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)?
- PMF:
    - A Probability Mass Function (PMF) is a function that gives the probability that a discrete random variable takes on a specific value

- PDF:
    - PDF its represents the density of coutineoues random variable lying b/w a specific range of values.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define a discrete random variable X
X = np.array([1, 2, 3, 4, 5, 6])

# Define the PMF of X
pmf_X = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Plot the PMF of X
plt.bar(X, pmf_X)
plt.title("PMF of a fair six-sided die")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()

# Define a continuous random variable Y
Y = np.linspace(-3, 3, 1000)

# Define the PDF of a standard normal distribution
pdf_Y = 1/np.sqrt(2*np.pi) * np.exp(-Y**2/2)

# Plot the PDF of Y
plt.plot(Y, pdf_Y)
plt.title("PDF of a standard normal distribution")
plt.xlabel("Value")
plt.ylabel("Probability density")
plt.show()
```

#### What is the difference between Bernoulli Distribution and Binomial Distribution?
- The main difference between Bernoulli distribution and binomial distribution is that Bernoulli distribution models a single experiment with two possible outcomes, while binomial distribution models a series of independent Bernoulli trials with a fixed number of trials (n) and the same probability of success (p) for each trial. #### Q8: What is the z score? 
- A z-score, also known as a standard score, is a measure of how many standard deviations a data point is away from the mean of a distribution.  z = (x - μ) / σ
#### State the importance of the z score.
- The importance of the z-score lies in its ability to standardize data and allow for meaningful comparisons between data points from different distributions. 
- By calculating the z-score, we can determine how far a data point is from the mean relative to the spread of the distribution. A positive z-score indicates that the data point is above the mean, while a negative z-score indicates that the data point is below the mean.

#### Q9: What is Central Limit Theorem? 
- we take repeated random samples from a population and calculate the mean of each sample, those means will be approximately normally distributed, with a mean equal to the population mean and a standard deviation equal to the population standard deviation divided by the square root of the sample size.
#### State the significance of the Central Limit Theorem.
- It provides a theoretical foundation for many statistical techniques, including hypothesis testing, confidence intervals, and parameter estimation.

### Q10: State the assumptions of the Central Limit Theorem.

Independence: The sample observations must be independent of each other. That is, the value of one observation should not be influenced by the values of the other observations.

Sample size: The sample size should be large enough. There is no exact rule for determining the minimum sample size required for the CLT to hold, but as a general guideline, a sample size of at least 30 is often considered large enough.

Population distribution: The population distribution should be reasonably well-behaved. Specifically, it should have a finite mean and a finite variance. If the population distribution is severely skewed or has heavy tails, the CLT may not hold even for large sample sizes.

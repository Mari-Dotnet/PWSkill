## check noraml distribution with QQ plotline

```python
import scipy.stats as stat
import pylab
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset('iris')
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
def load_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
    
```

```python
load_data(data,'sepal_width')
```

```python
load_data(data,'petal_width')
```

```python
load_data(data,'sepal_length')
```

```python
load_data(data,'petal_length')
```

```python
data_tips=sns.load_dataset('tips')
```

```python
data_tips.head()
```

```python
load_data(data_tips,'total_bill')
```

```python
load_data(data_tips,'tip')
```

```python
load_data(data_tips,'size')
```

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. 
#### Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
fmri=sns.load_dataset('fmri')
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri, hue='event')
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri, hue='region')
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,data=fmri, hue='region',style='event')
```

#### Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column.

#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot: It shows the relationship between two variables using dots on a 2D plane. It is useful for identifying patterns and trends in data.

Line Plot: It is used to plot data points connected with straight lines. It is useful for visualizing trends over time.

Bar Plot: It is used to show the relationship between a categorical variable and a numerical variable. It is useful for comparing different categories.

Histogram: It is used to visualize the distribution of a single variable. It is useful for understanding the shape and spread of the data.

Heatmap: It is used to visualize the correlation between variables in a dataset. It is useful for identifying relationships and patterns in complex datasets.

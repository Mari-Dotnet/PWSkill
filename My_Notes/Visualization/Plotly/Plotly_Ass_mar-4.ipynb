#### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import plotly.express as px
import seaborn as sns
```

```python
data=sns.load_dataset('titanic')
data.head()
```

```python
fig1=px.scatter(data, x='age',y='fare',title='Age vs Fare')
fig1.show()
```

#### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
data2=px.data.tips()
data2.head()
```

#### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column inthe tips dataset.  Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
data2.head()
```

```python
fig3=px.histogram(data2,x='sex',y='total_bill',pattern_shape='smoker',color='day')
fig3.show()
```

#### Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column forthe color parameter.

```python
data4=px.data.iris()
data4.head()
data4[data4.dtypes[data4.dtypes=='float64'].index]
```

```python
fig4=px.scatter_matrix(data4[data4.dtypes[data4.dtypes=='float64'].index],color=data4['species'])
fig4.show()
```

### Q5. What is Distplot? 
- Combined statistical representations with px.histogram

### Using Plotly express, plot a distplot.

```python
import plotly.express as px
df = px.data.tips()
fig = px.histogram(df, x="total_bill", y="tip", color="sex",
                   marginal="box", # or violin, rug
                   hover_data=df.columns)
fig.show()
```

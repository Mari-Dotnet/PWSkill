#### Q2: What is a scatter plot?
- scatter plot used to plot the data point on horizandal  and vertical axis

```python
#  Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y, c='r')
plt.xlabel("This is my x labe")
plt.ylabel("this is my y label")
plt.title("this is my title")
plt.grid()
```

#### Q4: What is a bar plot?
- bar plot is graphical representation of the categrical data using retangular bars with lenth propotional to the values present.
#### Why is it used? 
- Bar plots are commonly used to compare the values of different categories

```python
## Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color="blue")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Company vs Profit")
```

#### Q5: What is a box plot? 
- box plot is a graphical representation of the distribution of numerical data through their quartiles.
#### Why is it used?
-  Box plots are useful for identifying and visualizing the distribution, skewness, and outliers in a data set. They are commonly used in data analysis, statistics, and machine learning.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data=[box1,box2]
```

```python
fig,ax=plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_title('Box Plot')
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
- To create multiple plot in same figure.

```python

import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2)
ax[0,0].plot(x1,y1)
ax[0, 0].set_title('Line 1')

ax[0,1].bar(x2,y2)
ax[0, 1].set_title('Line 2')

ax[1,0].scatter(x3,y3)
ax[1,0].set_title('Line 3')

ax[1,1].plot(x4,y4)
ax[1, 1].set_title('Line 3')

plt.show()
```

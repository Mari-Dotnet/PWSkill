#### Q5. What function is used in Flask for URL Building? 
- url_for function is used for to generate the URL 

```python
### Write a Python code to demonstrate the working of the url_for() function.
```

```python
from flask import Flask,url_for,redirect


app = Flask(__name__)

@app.route("/")
def hello_world():
   return redirect(url_for('username',name ='mari'))


@app.route('/user/<name>')
def username(name):
    return "hello user {}".format(name)


if __name__=="__main__":
    app.run(host="0.0.0.0")
```
